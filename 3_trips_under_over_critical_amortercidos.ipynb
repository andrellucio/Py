```python
import control
import matplotlib.pyplot as plt
```

```python
# Define transfer function of the system
s = control.tf('s')
```

## Process  - Second Order System

A second-order linear system is a common description of many dynamic processes. The response depends on whether it is an overdamped, critically damped, or underdamped second order system.

```python
#Natural Period of Oscillation
Tau = 200

#Process Gain
K = 1.0
```

### Overdamped second order system

```python
Zeta = 1.5

A = Tau**2
B = 2*Zeta*Tau
C = 1
sys_over = K/(A*s**2 + B*s + 1)
```

```python
# Simulate step response
t_over, y_over = control.step_response(sys_over, 2500)
```

### Critically damped second order system

```python
Zeta = 1.0

A = Tau**2
B = 2*Zeta*Tau
C = 1
sys_cri = K/(A*s**2 + B*s + 1)
```

```python
# Simulate step response
t_cri, y_cri = control.step_response(sys_cri, 2500)
```

### Underdamped second order system

```python
Zeta = 0.01

A = Tau**2
B = 2*Zeta*Tau
C = 1
sys_und = K/(A*s**2 + B*s + 1)
```

```python
# Simulate step response
t_und, y_und = control.step_response(sys_und, 2500)
```

## Plots

```python
# Plot step response
plt.plot(t_und, y_und, label="Underdamped")
plt.plot(t_cri, y_cri, label="Critically damped")
plt.plot(t_over, y_over, label="Overdamped")

plt.xlabel('Time (s)')
plt.ylabel('Output')
plt.title('System Step Response')
plt.grid()
plt.legend(loc="best")
plt.show()
```
